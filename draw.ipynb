```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./result.csv")
df
```

```python
color_list = ['black', 'red', 'gold', 'green', 'blue', 'silver', 'pink', 'peru', 'lawngreen', 'cyan', 'darkviolet']
shape_list = ['o', 'P', '^','s', '*', 'X', 'D']
plt.xlabel("thread num")
plt.ylabel("cost time/ms")
x = np.linspace(1,10,10)
for index, row in df.iterrows():
    # print(index)
    # print(row)
    y = []
    for i in range(1,11):
        y.append(row[i])
    print(y)
    plt.plot(x,y,color=color_list[index], marker =shape_list[index], label=row["TestName"])
plt.legend()
plt.savefig("./result.png",facecolor='w')
```
